```python
#reading file without a header and telling it not to put header
df = pandas.read_csv("supermarkets-commas - Copy.txt", header=None)
# pandas.read_csv?
df
```

```python
#manual header/column definition on the file above 
df.columns=["ID","Address", "City", "ZIP", "Country","Name", "Employee"]
df
```

```python
#setting a row
rowSet = df.set_index("ID") #this makes a copy of the above making the ID as the row value
rowSet
```

```python
df #this will print the code before the above
```

```python
#another way to make it permanent without having to change the name
df.set_index("ID", inplace=True) #set id as index, take it in place, yes
df
```

```python
#setting another column as index will by default drop the initial one, ID, set
df.set_index("Address", inplace=True, drop=False) #take address as index, take it to place, but dont drop
df
```
